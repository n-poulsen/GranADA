```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("../data/Production_Crops_E_All_Data/Production_Crops_E_All_Data.csv",encoding = 'unicode_escape')
```

```python
dfar=df[df["Element"]=="Area harvested"]
data = dfar.set_index("Area")
data = data.drop(["World","Asia","Net Food Importing Developing Countries","Americas",\
                 "Low Income Food Deficit Countries","Africa","Land Locked Developing Countries",\
                 "South America","Eastern Asia","Least Developed Countries","Europe","European Union",\
                 "Central America","Southern Europe","Oceania","China","Western Asia","Southern Asia",\
                 "Eastern Africa","Northern Africa","Eastern Europe","South-Eastern Asia","Northern America",\
                 "Small Island Developing States","Western Europe","Central Asia","Western Africa","Middle Africa",\
                 "Caribbean","Southern Africa"], axis=0)
data.reset_index(level=0, inplace=True)
dfar=data
```

```python
for a in dfar["Area"].value_counts().index:
    print(a,dfar["Area"].value_counts()[a])
```

```python
dfent=pd.DataFrame()
ent1=0

for item in dfar.Item.unique():

    entsr = []
    dfban=dfar[dfar["Item"]==item]
    for yr in range(1961,2018):
        yr1="Y"+str(yr)
        p1=dfban[yr1]/dfban[yr1].sum()
        ent=p1*np.log2(p1)*-1
        if ent.sum()>0:
            entsr.append(ent.sum())
        else:
            entsr.append(np.NaN)
        
        ent2=ent.sum()
        entdif=np.abs(ent2-ent1)
        if entdif<1 and entdif>0.5 and yr>1961:
            print(item,yr)
        ent1=ent.sum()

    dfent[item] = entsr

```

```python
dfent.plot(legend=False)
plt.show()
dfent.describe()
```

```python
plt.plot(dfent.mean(axis=1).values)
plt.xlabel("years")
plt.ylabel("Entropy average")
```

```python
dfentvar={}
varlst=[]
for item in dfent.columns:
    dfentvar[item]=dfent[item].var()
    if dfent[item].var()<2.5 and dfent[item].var()>0.3:
        print(item,"\t\t",dfentvar[item])
        varlst.append(item)

dfent[varlst].plot(legend=False)
```
